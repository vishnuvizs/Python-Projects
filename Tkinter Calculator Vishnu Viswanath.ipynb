```python
from tkinter import *
from tkinter import ttk
import math
import parser
```

```python
window = Tk()
window.title("Calculator")
window.geometry("250x250")
mymenu= Menu(window)
window.config(menu=mymenu)
file = Menu(mymenu, tearoff = 0)
mymenu.add_cascade(label = "File", menu = file)

tab_control = ttk.Notebook(window)

tab1 = ttk.Frame(tab_control)
tab2 = ttk.Frame(tab_control)

tab_control.add(tab1, text='Normal')
tab_control.add(tab2, text='Scientific')

display = Entry(tab1, bd = 5)
display1 = Entry(tab2, bd = 5)
display.grid(row= 2, columnspan = 6, sticky = W+E)
display1.grid(row= 2, columnspan = 6, sticky = W+E)


i = 0
def get_num(num):
    global i
    display.insert(i,num)
    i +=1

def clear_all():
    display.delete(0, END)
    
def clear_one():
    string = display.get()
    if display.get() == "Error":
        clear_all()
    elif len(string):
        a = string[:-1]
        clear_all()
        display.insert(0,a)
    else:
        clear_all()
        display.insert(0, "Error")
        
def get_operator(oper):
    global i
    length = len(oper)
    display.insert(i,oper)
    i += length
    
def equal_to():
    string = display.get()
    try:
        a = parser.expr(string).compile()
        result = eval(a)
        clear_all()
        display.insert(0,result)
    except:
        clear_all()
        display.insert(0,"Error")
        
        
        
j = 0
def get_num1(num):
    global j
    display1.insert(j,num)
    k = len(display1.get())
    j = k +1

def clear_all1():
    display1.delete(0, END)
    
def clear_one1():
    string = display1.get()
    if display1.get() == "Error":
        clear_all1()
    elif len(string):
        a = string[:-1]
        clear_all1()
        display1.insert(0,a)
    else:
        clear_all1()
        display1.insert(0, "Error")
        
def get_operator1(oper):
    global j
    length = len(oper)
    display1.insert(j,oper)
    j += length
    
def equal_to1():
    string = display1.get()
    try:
        if "sin(" in string:
            b = string.find('sin(')
            e = b + 4
            c = string.find(')')
            d = int(string[e : c])
            a = math.radians(d)
            result = math.sin(a)
            clear_all1()
            display1.insert(0,result)
        else:
            a = parser.expr(string).compile()
            result = eval(a)
            clear_all1()
            display1.insert(0,result)
    except:
        clear_all1()
        display1.insert(0,"Error")
        

        
        
Button(tab1, text ="1", command = lambda :get_num(1), width = 3).grid(row=4,column=0)
Button(tab1, text ="2", command = lambda :get_num(2), width = 3).grid(row=4,column=1)
Button(tab1, text ="3", command = lambda :get_num(3), width = 3).grid(row=4,column=2)
Button(tab1, text ="4", command = lambda :get_num(4), width = 3).grid(row=5,column=0)
Button(tab1, text ="5", command = lambda :get_num(5), width = 3).grid(row=5,column=1)
Button(tab1, text ="6", command = lambda :get_num(6), width = 3).grid(row=5,column=2)
Button(tab1, text ="7", command = lambda :get_num(7), width = 3).grid(row=6,column=0)
Button(tab1, text ="8", command = lambda :get_num(8), width = 3).grid(row=6,column=1)
Button(tab1, text ="9", command = lambda :get_num(9), width = 3).grid(row=6,column=2)
Button(tab1, text ="0", command = lambda :get_num(0), width = 3).grid(row=7,column=1)

Button(tab1, text ="^", command = lambda :get_operator("**"), width = 3).grid(row=3,column=0)
Button(tab1, text ="√", command = lambda :get_operator("√"), width = 3).grid(row=3,column=1)
Button(tab1, text ="del",command = lambda :clear_one(), width = 3).grid(row=3,column=2)
Button(tab1, text ="C", command = lambda :clear_all(), width = 3).grid(row=3,column=3)
Button(tab1, text ="+", command = lambda :get_num("+"), width = 3).grid(row=4,column=3)
Button(tab1, text ="-", command = lambda :get_num("-"), width = 3).grid(row=5,column=3)
Button(tab1, text ="/", command = lambda :get_num("/"), width = 3).grid(row=6,column=3)
Button(tab1, text =".", command = lambda :get_num("."), width = 3).grid(row=7,column=0)
Button(tab1, text ="=", command = lambda :equal_to(), width = 3).grid(row=7,column=2)
Button(tab1, text ="x", command = lambda :get_num("*"), width = 3).grid(row=7,column=3)


Button(tab2, text ="1", command = lambda :get_num1(1), width = 3).grid(row=5,column=0)
Button(tab2, text ="2", command = lambda :get_num1(2), width = 3).grid(row=5,column=1)
Button(tab2, text ="3", command = lambda :get_num1(3), width = 3).grid(row=5,column=2)
Button(tab2, text ="4", command = lambda :get_num1(4), width = 3).grid(row=6,column=0)
Button(tab2, text ="5", command = lambda :get_num1(5), width = 3).grid(row=6,column=1)
Button(tab2, text ="6", command = lambda :get_num1(6), width = 3).grid(row=6,column=2)
Button(tab2, text ="7", command = lambda :get_num1(7), width = 3).grid(row=7,column=0)
Button(tab2, text ="8", command = lambda :get_num1(8), width = 3).grid(row=7,column=1)
Button(tab2, text ="9", command = lambda :get_num1(9), width = 3).grid(row=7,column=2)
Button(tab2, text ="0", command = lambda :get_num1(0), width = 3).grid(row=8,column=1)

Button(tab2, text ="^", command = lambda :get_operator1("**"), width = 3).grid(row=4,column=0)
Button(tab2, text ="√", command = lambda :get_operator1("√"), width = 3).grid(row=4,column=1) 
Button(tab2, text ="del",command = lambda :clear_one1(), width = 3).grid(row=4,column=2)
Button(tab2, text ="C", command = lambda :clear_all1(), width = 3).grid(row=4,column=3)
Button(tab2, text ="+", command = lambda :get_num1("+"), width = 3).grid(row=5,column=3)
Button(tab2, text ="-", command = lambda :get_num1("-"), width = 3).grid(row=6,column=3)
Button(tab2, text ="/", command = lambda :get_num1("/"), width = 3).grid(row=7,column=3)
Button(tab2, text =".", command = lambda :get_num1("."), width = 3).grid(row=8,column=0)
Button(tab2, text ="=", command = lambda :equal_to1(), width = 3).grid(row=8,column=2)
Button(tab2, text ="x", command = lambda :get_num1("*"), width = 3).grid(row=8,column=3)
Button(tab2, text ="(", command = lambda :get_num1("("), width = 3).grid(row=4,column=4)
Button(tab2, text =")", command = lambda :get_num1(")"), width = 3).grid(row=5,column=4)
Button(tab2, text ="sin", command = lambda :get_num1("sin("), width = 3).grid(row=6,column=4)
Button(tab2, text ="cos", command = lambda :get_num1("sin("), width = 3).grid(row=7,column=4)
Button(tab2, text ="tan", command = lambda :get_num1("sin("), width = 3).grid(row=8,column=4)
Button(tab2, text ="π", command = lambda :get_num1(1), width = 3).grid(row=3,column=0)
Button(tab2, text ="!", command = lambda :get_num1(1), width = 3).grid(row=3,column=1)
Button(tab2, text ="Exp", command = lambda :get_num1(1), width = 3).grid(row=3,column=2)
Button(tab2, text ="^10", command = lambda :get_num1(1), width = 3).grid(row=3,column=3)
Button(tab2, text ="1", command = lambda :get_num1(1), width = 3).grid(row=3,column=4)

tab_control.grid(row=1,column = 0)

file.add_command(label ="New",command = clear_all)
file.add_command(label ="Save")
file.add_separator()
file.add_command(label ="Quit", command = window.destroy)

window.mainloop()
```
